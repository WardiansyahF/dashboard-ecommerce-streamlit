### Cleaning Data

#### Order Data

Mengubah data date pada data order yang sebelumnya object menjadi datetime

```python
date_column = ['order_delivered_carrier_date','order_delivered_customer_date','order_estimated_delivery_date']

for column in date_column :
  orders_df[column] = pd.to_datetime(orders_df[column])

orders_df.info()
```

drop value NaN pada dataset order sehingga tidak mengganggu proses analisis

```python
orders_df.dropna(axis = 0,inplace = True)

orders_df.isna().sum()
```

#### Products Data

Men-drop nilai NaN pada dataset Products sehingga tidak mengganggu proses analisis

```python
products_df.dropna(axis = 0,inplace = True)

products_df.isna().sum()
```

## Exploratory Data Analysis (EDA)

### Merging Products Data dan Orders Data

```python
order_items_products_df = pd.merge(
    left=order_items_df,
    right=products_df,
    how="left",
    left_on="product_id",
    right_on="product_id"
)
order_items_products_df.head()
```

Melihat berapa banyak kategori produk yang terjual dan berapa revenue yang dihasilkan dari total penjualan produk tersebut

```python
order_items_products_df.groupby(by="product_category_name").agg({
    "order_id" : "nunique",
    "price" : "sum"
})
```

```python
order_items_products_df.groupby(by="product_category_name").order_id.nunique().sort_values(ascending=False)
```

dapat disimpulkan bahwa kategori cama_mesa_banho merupakan kategori produk yang paling banyak terjual

```python
order_items_products_df.groupby(by="product_category_name").price.sum().sort_values(ascending=False)
```

dapat disimpulkan bahwa kategori beleza_saude merupakan produk yang menyumbang revenue paling banyak

### Active Customers & Non Active Customers

Membuat kolom status dimana jika customer_id terdapat pada dataset order, maka status customer adalah Active, jika tidak maka Non Active.
Active menandakan customer pernah melakukan pembelian setidaknya sekali, sedangkan non active berarti tidak pernah sama sekali.

```python
customer_id_in_orders_df =  orders_df.customer_id.tolist()
customers_df["status"] = customers_df["customer_id"].apply(lambda x: "Active" if x in customer_id_in_orders_df else "Non Active")
customers_df.sample(25)
```

Terlihat bahwa status telah diterapkan pada dataset customers, sekarang kita akan menghitung berapa banyak status active dan non active-nya

```python
customers_df.groupby(by="status").customer_id.count()
```

Terlihat bahwa pembeli aktif jauh lebih banyak dibandingkan pembeli non active. sekarang kita melihat state yang memiliki customer active paling banyak

```python
active_customers_df = customers_df.loc[customers_df['status'] == 'Active']
active_customers_df.groupby(by="customer_state").status.count().sort_values(ascending=False).head(10)
```

```python
active_customers_df = customers_df.loc[customers_df['status'] == 'Non Active']
active_customers_df.groupby(by="customer_state").status.count().sort_values(ascending=False).head(10)
```

Terlihat bahwa Sao Paulo(SP) merupakan state yang memiliki customer active paling banyak dan juga merupakan stete yang memiliki customer Non active paling banyak

## Visualization & Explanatory Analysis

### Pertanyaan 1: Kategori barang apa yang paling banyak diminati/dibeli dan paling sedikit diminati/dibeli?

```python
product_category_counts = order_items_products_df.groupby(by="product_category_name").order_id.nunique().sort_values()

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15, 6))

axes[0].bar(product_category_counts.tail(10).index[::-1], product_category_counts.tail(10).values[::-1], color='skyblue')
axes[0].set_title('Top 10 Most Ordered Product Categories')
axes[0].set_xlabel('Product Category')
axes[0].set_ylabel('Number of Orders')
axes[0].tick_params(axis='x', rotation=45)

axes[1].bar(product_category_counts.head(10).index, product_category_counts.head(10).values, color='lightcoral')
axes[1].set_title('Top 10 Least Ordered Product Categories')
axes[1].set_xlabel('Product Category')
axes[1].set_ylabel('Number of Orders')
axes[1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

Terlihat bahwa pada Top 10 kategori produk yang dibeli, cama_mesa_banho menjadi kategori produk yang paling banyak dibeli

### Pertanyaan 2: Kategori barang mana yang menyumbang revenue paling banyak terhadap perusahaan?

```python
product_category_prices = order_items_products_df.groupby(by="product_category_name").price.sum().sort_values(ascending=False)


plt.figure(figsize=(10, 6))
product_category_prices.head(10).plot(kind='bar', color='lightgreen')


plt.title('Top 10 Product Categories by Total Price')
plt.xlabel('Product Category')
plt.ylabel('Total Price')


plt.show()
```

Pada visualisasi diatas kita mendapatkan data Top 10 penyumbang revenue terbesar, didapatkan beleza_saude sebagai kategori produk dengan penyumbang revenue terbesar

### Pertanyaan 3: State mana yang memiliki pembeli active paling banyak?

```python
active_customers_df = customers_df.loc[customers_df['status'] == 'Active']
active_customer_states = active_customers_df.groupby(by="customer_state").status.count().sort_values(ascending=False).head(10)


non_active_customers_df = customers_df.loc[customers_df['status'] == 'Non Active']
non_active_customer_states = non_active_customers_df.groupby(by="customer_state").status.count().sort_values(ascending=False).head(10)


fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15, 6))


axes[0].bar(active_customer_states.index, active_customer_states.values, color='lightblue')
axes[0].set_title('Top 10 States with Active Customers')
axes[0].set_xlabel('Customer State')
axes[0].set_ylabel('Number of Active Customers')
axes[0].tick_params(axis='x', rotation=45)


axes[1].bar(non_active_customer_states.index, non_active_customer_states.values, color='lightcoral')
axes[1].set_title('Top 10 States with Non-Active Customers')
axes[1].set_xlabel('Customer State')
axes[1].set_ylabel('Number of Non-Active Customers')
axes[1].tick_params(axis='x', rotation=45)


plt.tight_layout()
plt.show()
```

dari visualisasi diatas dikategorikan menjadi Top 10 Most Active Users dan Top 10 Most Non Active Users didapati bahwa kode state SP menjadi yang terbesar baik Active maupun Non Active usersnya
